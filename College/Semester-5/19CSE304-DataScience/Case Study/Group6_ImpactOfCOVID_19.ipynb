```python
dfv.fillna(value = 0, inplace = True)
dfv.total_vaccinations = dfv.total_vaccinations.astype(int)
dfv.people_vaccinated = dfv.people_vaccinated.astype(int)
dfv.people_fully_vaccinated = dfv.people_fully_vaccinated.astype(int)

dfv.daily_vaccinations_raw = dfv.daily_vaccinations_raw.astype(int)
dfv.daily_vaccinations = dfv.daily_vaccinations.astype(int)
dfv.total_vaccinations_per_hundred = dfv.total_vaccinations_per_hundred.astype(int)

dfv.people_fully_vaccinated_per_hundred = dfv.people_fully_vaccinated_per_hundred.astype(int)
dfv.daily_vaccinations_per_million = dfv.daily_vaccinations_per_million.astype(int)
dfv.people_vaccinated_per_hundred = dfv.people_vaccinated_per_hundred.astype(int)
```

```python
dfv.country.unique()
```

```python
date = dfv.date.str.split('-', expand =True)
dfv['year'] = date[0]
dfv['month'] = date[1]
dfv['day'] = date[2]

dfv.year = pd.to_numeric(dfv.year)
dfv.month = pd.to_numeric(dfv.month)
dfv.day = pd.to_numeric(dfv.day)
dfv.date = pd.to_datetime(dfv.date)
```

```python
print('Total no of countries in the data set ',len(dfv.country.unique()),'\n')
print('Total no of unique vaccines in the data set ',len(dfv.vaccines.unique()),'\n')
```

```python
print('Data point starts from ',dfv.date.min(),'\n')
print('Data point ends at ',dfv.date.max(),'\n')
```

```python
country_wise_total_vaccinated = {}
for country in dfv.country.unique() : 
    vaccinated = 0
    for i in range(len(dfv)) : 
        if dfv.country[i] == country : 
            vaccinated += dfv.daily_vaccinations[i]
    
    country_wise_total_vaccinated[country] = vaccinated 
#     made a seperate dict from the df 
    tv_df = pd.DataFrame.from_dict(country_wise_total_vaccinated,orient='index',columns = ['total_vaccinted_till_date'])
#     converted dict to df 
tv_df.sort_values(by = 'total_vaccinted_till_date', ascending = False, inplace = True)
tv_df
```
