```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("https://raw.githubusercontent.com/getup8/datasets/master/2019_Car_Fuel_Economy.csv")
```

```python
data.dtypes
```

```python
data.describe()
```

```python
data.head()
```

```python
data.shape
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(x='Air Aspiration Method Desc',y='Cylinders',data=data)
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(x='Air Aspiration Method Desc',y='Cylinders',data=data,inner='quartile')
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(x='Air Aspiration Method Desc',y='Cylinders',data=data,inner=None)
```

```python
plt.figure(figsize=(8,8))
sns.violinplot(y='Air Aspiration Method Desc',x='Cylinders',data=data)
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(x='Air Aspiration Method Desc',y='Cylinders',data=data,inner=None,cut=3,scale='count')
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(x='Air Aspiration Method Desc',y='Cylinders',data=data,cut=3,bw=0.5)
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(x='Air Aspiration Method Desc',y='Cylinders',data=data,inner=None,hue='Trans')
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(x='Air Aspiration Method Desc',y='Cylinders',data=data,linewidth=3)
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(x='Air Aspiration Method Desc',y='Cylinders',data=data,palette='pastel')
```

```python
plt.figure(figsize=(10,10))
sns.stripplot(x='Trans', y="Displacement", data=data, jitter=0.03)
sns.violinplot(x='Trans',y='Displacement',data=data)
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(x='Trans',y='Cylinders',data=data,saturation=0.2)
```

```python
plt.figure(figsize=(10,10))
sns.violinplot(x='Trans',y='Cylinders',data=data,color='blue')
```

```python
plt.figure(figsize=(10,10))
sns.swarmplot(x='Air Aspiration Method Desc', y="Displacement", data=data)
sns.violinplot(x='Air Aspiration Method Desc',y='Displacement',data=data)
```

```python
plt.figure(figsize=(10,10))
sns.swarmplot(x='Air Aspiration Method Desc', y="Displacement", data=data,palette='pastel')
sns.violinplot(x='Air Aspiration Method Desc',y='Displacement',data=data,palette='muted')
```

```python
plt.figure(figsize=(10,10))
sns.swarmplot(x='Air Aspiration Method Desc', y="Displacement", data=data,linewidth=2)
sns.violinplot(x='Air Aspiration Method Desc',y='Displacement',data=data)
```

```python
plt.figure(figsize=(10,10))
sns.swarmplot(x='Air Aspiration Method Desc', y="Displacement",data=data,linewidth=2)
sns.violinplot(x='Air Aspiration Method Desc',y='Displacement',data=data,inner=None)
```

```python
plt.figure(figsize=(60,30))
sns.stripplot(x='Air Aspiration Method Desc',y='# Gears',data=data,linewidth=1)
sns.violinplot(x='Air Aspiration Method Desc',y='# Gears',data=data)
```

```python
plt.figure(figsize=(60,30))
sns.stripplot(x='Air Aspiration Method Desc',y='# Gears',data=data,linewidth=1)
sns.violinplot(x='Air Aspiration Method Desc',y='# Gears',data=data,hue='Trans')
```

```python
plt.figure(figsize=(60,30))
sns.stripplot(x='Air Aspiration Method Desc',y='# Gears',data=data,linewidth=2,palette='bright')
sns.violinplot(x='Air Aspiration Method Desc',y='# Gears',data=data,palette='muted',inner=None)
```

```python
plt.figure(figsize=(10,10))
sns.stripplot(x='# Gears',y='City Unadj FE - Conventional Fuel',data=data,linewidth=1)
sns.violinplot(x='# Gears',y='City Unadj FE - Conventional Fuel',data=data,inner='quartile')
```
