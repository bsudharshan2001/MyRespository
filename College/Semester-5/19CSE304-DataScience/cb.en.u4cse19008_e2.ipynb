```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```
columns = ['Age','Workclass','fnlgwt','Education','Education num','Marital Status',
           'Occupation','Relationship','Race','Sex','Capital Gain','Capital Loss',
           'Hours/Week','Native country','Income']
train = pd.read_csv('https://raw.githubusercontent.com/vivek2319/Predicting-US-Census-Income/master/Dataset/adult-training.csv', names=columns)
```

```
train.describe()
```

```
train.dtypes
```

# Q1

```
sns.scatterplot(x='Capital Loss',y='Capital Gain',data=train,hue='Marital Status',size='Age')
```

Inference: The Scatter plot graph depicts the how the Capital gain and Capital Loss is Scatterd based on their Marital Status and age

# Q2

```
sns.lineplot(x="Age", y="Hours/Week",color='green',data=train,marker='<')
```

Inference: The graph show the relation between Hours/Week and Age (How they vary with respect to each other). As the age increases, A person's work hours per week decreases.

# Q3

```
sns.stripplot(x='Age', y='Marital Status', data=train)
sns.boxplot(x='Age', y='Marital Status', data=train)
```

Inference: The graph plot shows the boxplot as well as strip plot between Age and Martital Status of each Age Groups



# Q4

```
plt.figure(figsize=(10,10))
sns.stripplot(x='Relationship', y='Age',data=train,jitter=0.3)
sns.violinplot(x='Relationship', y='Age',data=train)
```

Inference: The graph shows the scatter plot as well as the violin plot of how Relationship differs with age

# Q5

```
plt.figure(figsize=(10,10))
sns.boxplot(x='Marital Status', y='Age', data=train, hue='Sex')
plt.legend(bbox_to_anchor=(1.1, 0.52), loc='upper left', borderaxespad=0)
```

Inference: This Box plot show how the data is wrt to Age and Marital status of each person with Sex as indicator

# Q6

```
plt.figure(figsize=(10,10))
sns.violinplot(y='Marital Status',x='Age',data=train,linewidth=2,hue='Sex',split = True,bw = 0.1)
```

Inference: This graph shows the violin plot of Age Groups and their marital Status as shown with increase in age the no of widows increase

# Q7

```
fig=sns.barplot(x=train['Income'],y=train['Age'],hue=train['Sex'],dodge=False)
plt.title('Age vs Income',color='white',size=30)
plt.legend(bbox_to_anchor=[1.2, 0.5], loc='center');
plt.show()
```

Inference: This is a barplot of income wrt to Age with Sex as Indicator

# Q8

```
sns.catplot(x="Age", y="Marital Status", hue="Income",col="Sex", aspect=.6,kind="box", data=train);
```

Inference: The graph depicts the age outliers for both male and female, based on marital status and income which is divided into two brackets (less than 50k and greater than 50k).
